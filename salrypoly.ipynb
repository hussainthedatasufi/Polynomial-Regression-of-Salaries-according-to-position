```python
#Seperating Input and output
x=df.iloc[:,1:2].values
y=df.iloc[:,2].values
x
```

```python
#there will be no Splitting the dataset into the Training set and Test set
#there will be no feature scaling
```

# Model Creation

```python
#Training the linear regression from whole dataset
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)
```

```python
#Training the Polynomial regression of degree 2 from whole dataset
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
reg2=LinearRegression()
reg2.fit(x_poly,y)
```

```python
#Training the polynomial regression of degree 3 from whole dataset
```

```python
poly3=PolynomialFeatures(degree=3)
x_poly3=poly3.fit_transform(x)
reg3=LinearRegression()
reg3.fit(x_poly3,y)
```

```python
#Training the polynomial regression of degree 4 from whole dataset
```

```python
poly4=PolynomialFeatures(degree=4)
x_poly4=poly4.fit_transform(x)
reg4=LinearRegression()
reg4.fit(x_poly4,y)
```

```python
#Visualising the Linear Regression Result
```

```python
plt.rcParams["figure.figsize"] = (15,7)
plt.scatter(x,y,color='black')
plt.plot(x,reg.predict(x),color='red')
plt.title("Linear Regression")
plt.xlabel('Position level')
plt.ylabel('Salary')
```

```python
#Visualising the Polynomial Regression result for 2 degree
```

```python
plt.scatter(x,y,color='black')
plt.plot(x,reg2.predict(x_poly),color='blue')
plt.title("Polynomial Regression degree 2")
plt.xlabel('Position level')
plt.ylabel('Salary')
```

```python
#Visualising Polynomial regression with degree 3
```

```python
plt.scatter(x,y,color='black')
plt.plot(x,reg3.predict(x_poly3),color='aqua')
plt.title("Polynomial Regression degree 3")
plt.xlabel('Position level')
plt.ylabel('Salary')
```

```python
#Visulaising the result of Polynomial Regression of degree 4
```

```python
plt.scatter(x,y,color='black')
plt.plot(x,reg4.predict(x_poly4),color='green')
plt.title("Polynomial Regression degree 4")
plt.xlabel('Position level')
plt.ylabel('Salary')
```

```python
#Comparing all results
```

```python
plt.scatter(x,y,color='black')
plt.plot(x,reg.predict(x),color='red')
plt.plot(x,reg2.predict(x_poly),color='blue')
plt.plot(x,reg3.predict(x_poly3),color='aqua')
plt.plot(x,reg4.predict(x_poly4),color='green')
plt.title("Polynomial Regression V/s Linear Regression")
plt.xlabel("Positions")
plt.ylabel("Salaries")
```
